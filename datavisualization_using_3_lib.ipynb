```python
# prompt: using sklearn,pytorch,tensorflow creare a data visulaization

# Install necessary libraries
!pip install scikit-learn matplotlib tensorflow torch torchvision torchaudio

# Import libraries
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import make_blobs
import tensorflow as tf
import torch

# Generate sample data using scikit-learn
X, y = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)

# Visualize the data using matplotlib
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='viridis')
plt.title('Data Visualization using Matplotlib')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()


# Tensorflow Visualization (Example: Histogram of feature values)
tf_X = tf.convert_to_tensor(X, dtype=tf.float32)

plt.hist(tf_X[:, 0].numpy(), bins=20, color='skyblue', edgecolor='black')
plt.title('TensorFlow Visualization: Histogram of Feature 1')
plt.xlabel('Feature 1 Values')
plt.ylabel('Frequency')
plt.show()


# PyTorch visualization (Example: Scatter plot using tensors)
torch_X = torch.tensor(X, dtype=torch.float32)
torch_y = torch.tensor(y, dtype=torch.long)


plt.scatter(torch_X[:, 0], torch_X[:, 1], c=torch_y, s=50, cmap='viridis')
plt.title('PyTorch Visualization: Scatter Plot')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```
